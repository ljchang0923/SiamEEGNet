```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
xls = pd.ExcelFile('within_subject_result.xlsx')
df1 = pd.read_excel(xls, 'RMSE')
df2 = pd.read_excel(xls, 'CC')
```

```python
colors = ['#087CD0','#0AC168', '#D8973C', '#D0052A']
df1.plot(x="Subject", y=["SVR", "SCCNet", "EEGNet", "Siamese SCCNet"], yerr = kind="bar", 
         width=0.6, edgecolor='k', linewidth=1.2, figsize=(12,6), color = colors)
plt.rc('font', size=15)
plt.rc('legend', fontsize=16)
plt.rc('axes', labelsize=20)
plt.xticks(rotation=0)
hfont = {'fontname':'Helvetica'}
plt.ylabel('RMSE', **hfont)
plt.legend(loc='upper right',ncol=2)
plt.ylim([0, 0.7])
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import ttest_ind

xls = pd.ExcelFile('cross_subject_result.xlsx')
df_ablation = pd.read_excel(xls, 'repeat')
# df_ablation = df_ablation.transpose()

repeat_result = df_ablation.to_numpy()
repeat_result = repeat_result[:,1:]
for i in range(2):
    res_rmse = ttest_ind(repeat_result[0, :], repeat_result[2 + 2*i, :])
    res_cc = ttest_ind(repeat_result[1, :], repeat_result[3 + 2*i, :])
    print(res_rmse)
    print(res_cc)
```
